```python
#define max and min rank for ordinal variable

maxRank = 3
minRank = 1
```

```python
#define max and min Age,Balance

maxAge = df['Age'].max()
minAge = df['Age'].min()
maxBalance = df['Balance'].max()
minBalance = df['Balance'].min()
```

```python
#define function which calculates dissimalirity between two ids

def calcDissim(id1,id2,dfr):
     #since numbering starts from 0 in the df, we substract 2 to find the corresponding ids
    id1 = id1 - 2
    id2 = id2 - 2
    #Age is numerical so d(a,b) = |a-b|/(max_value – min_value)
    dis1 = abs(dfr.iloc[id1].Age-dfr.iloc[id2].Age)/(maxAge-minAge)
    #Job is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Job == dfr.iloc[id2].Job:
        dis2 = 0
    else:
        dis2 = 1
    #Marital is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Marital == dfr.iloc[id2].Marital:
        dis3 = 0
    else:
        dis3 = 1
    #Education is ordinal so d(a,b) = |rank(a)- rank(b)|/(max_rank – min_rank)
    dis4 = abs(dfr.iloc[id1].Education-dfr.iloc[id2].Education)/(maxRank-minRank)
    #Default is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Default == dfr.iloc[id2].Default:
        dis5 = 0
    else:
        dis5 = 1
    #Balance is numerical so d(a,b) = |a-b|/(max_value – min_value)
    dis6 = abs(dfr.iloc[id1].Balance-dfr.iloc[id2].Balance)/(maxBalance-minBalance)
    #Housing is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Housing == dfr.iloc[id2].Housing:
        dis7 = 0
    else:
        dis7 = 1
    #Loan is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Loan == dfr.iloc[id2].Loan:
        dis8 = 0
    else:
        dis8 = 1
    #finally we return the average
    return (dis1+dis2+dis3+dis4+dis5+dis6+dis7+dis8)/8

#example dissimilarity between id=2 and id=5204
calcDissim(2,5204,df)    
```

```python
#define function which calculates weighted dissimalirity between two ids

def calcDissimWeighted(id1,id2,dfr):
     #since numbering starts from 0 in the df, we substract 2 to find the corresponding ids
    id1 = id1 - 2
    id2 = id2 - 2
    #define weight values for each variable
    wAge=2 
    wBalance=3
    wJob=2
    wMarital=1
    wLoan=3
    wHousing=3
    wEducation=1 
    wDefault=3
    #Age is numerical so d(a,b) = |a-b|/(max_value – min_value)
    dis1 = abs(dfr.iloc[id1].Age-dfr.iloc[id2].Age)/(maxAge-minAge)
    #Job is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Job == dfr.iloc[id2].Job:
        dis2 = 0
    else:
        dis2 = 1
    #Marital is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Marital == dfr.iloc[id2].Marital:
        dis3 = 0
    else:
        dis3 = 1
    #Education is ordinal so d(a,b) = |rank(a)- rank(b)|/(max_rank – min_rank)
    dis4 = abs(dfr.iloc[id1].Education-dfr.iloc[id2].Education)/(maxRank-minRank)
    #Default is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Default == dfr.iloc[id2].Default:
        dis5 = 0
    else:
        dis5 = 1
    #Balance is numerical so d(a,b) = |a-b|/(max_value – min_value)
    dis6 = abs(dfr.iloc[id1].Balance-dfr.iloc[id2].Balance)/(maxBalance-minBalance)
    #Housing is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Housing == dfr.iloc[id2].Housing:
        dis7 = 0
    else:
        dis7 = 1
    #Loan is categorical so d(a,b) = 1 if a≠b, 0 otherwise
    if dfr.iloc[id1].Loan == dfr.iloc[id2].Loan:
        dis8 = 0
    else:
        dis8 = 1
    #finally we return the dissimilarity
    return (wAge*dis1+wJob*dis2+wMarital*dis3+wEducation*dis4+wDefault*dis5+wBalance*dis6+wHousing*dis7+wLoan*dis8)/18

#example weighted dissimilarity between id=2 and id=5204 
calcDissimWeighted(2,5204,df)
```

```python
#create function for 10 nearest neighbours

def calc10nn(id1,dfr):
    #we define a dictionary to store our calculations
    aDict = {}
    #iterate over the dataframe
    for index, row in dfr.iterrows():
        #condition ton not include self in calculations
        if index != id1:
            #apply function from previous step to calculate dissimilarity
            aDict[index] = calcDissim(id1,index,dfr)
    #create a sorted list(assending)
    sorted_d = sorted(aDict.items(), key=operator.itemgetter(1))
    #take only top 10 items
    sorted_dTop10 = sorted_d[:10]
    #insert column names
    sorted_dTop10.insert(0,('id','dissimilarity'))
    return sorted_dTop10

#examples for ids=1230,5032,10001,24035,28948,35099,37693,39543,40002,42192
outList1230 = calc10nn(1230,df)
outList5032 = calc10nn(5032,df)
outList10001 = calc10nn(10001,df)
outList24035 = calc10nn(24035,df)
outList28948 = calc10nn(28948,df)
outList35099 = calc10nn(35099,df)
outList37693 = calc10nn(37693,df)
outList39543 = calc10nn(39543,df)
outList40002 = calc10nn(40002,df)
outList42192 = calc10nn(42192,df)

# formatting for our list
f = '{}|{}'

#print list using defined format
for i in outList1230:
    print(f.format(*i))
for i in outList5032:
    print(f.format(*i))
for i in outList10001:
    print(f.format(*i))
for i in outList24035:
    print(f.format(*i))
for i in outList28948:
    print(f.format(*i))
for i in outList35099:
    print(f.format(*i))
for i in outList37693:
    print(f.format(*i))
for i in outList39543:
    print(f.format(*i))
for i in outList40002:
    print(f.format(*i))
for i in outList42192:
    print(f.format(*i))
```
